```python
from langchain_openai import OpenAIEmbeddings
import numpy as np
from sklearn.cluster import KMeans
from search_result import search_result as docs
from models import ResultTemplate, SearchResult
```

```python
# 2. Create an embedding model instance
embeddings_model = OpenAIEmbeddings(model="text-embedding-3-large")

embeddings = embeddings_model.embed_documents(
    [f"{doc['title']} - {doc['snippet']}" for doc in docs]
)
```

```python
n_clusters = 7
kmeans = KMeans(random_state=42).fit(np.array(embeddings))

labels = kmeans.labels_

clusters = {}
for label, doc in zip(labels, docs):
    clusters.setdefault(label, []).append(doc)
```
